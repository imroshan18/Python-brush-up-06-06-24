```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample DataFrame with null values
data = {
    'A': [1, 2, np.nan, 4, 5, np.nan, 7, 8, 9, 10, np.nan, 12, 13, 14, 15, 16],
    'B': ['a', 'b', np.nan, 'd', 'e', 'f', 'g', np.nan, 'i', 'j', 'k', np.nan, 'm', 'n', 'o', 'p'],
    'C': [11, np.nan, 13, 14, 15, 16, np.nan, 18, 19, 20, 21, 22, 23, 24, np.nan, 26],
    'D': [21, 22, 23, 24, np.nan, 26, 27, np.nan, 29, 30, 31, 32, 33, 34, 35, 36],
    'E': [1.1, 2.2, 3.3, np.nan, 5.5, 6.6, 7.7, 8.8, 9.9, 10.1, 11.11, 12.12, np.nan, 14.14, 15.15, 16.16]
}
df = pd.DataFrame(data)

# Displaying the initial DataFrame
print("Initial DataFrame:")
display(df)

# Handling null values
# 1. Checking for null values
print("\nChecking for null values:")
print(df.isnull().sum())

# 2. Dropping rows with null values
df_dropna = df.dropna()
print("\nDataFrame after dropping rows with null values:")
display(df_dropna)

# 3. Filling null values with a specific value (e.g., mean, median)
df_filled_mean = df.fillna(df.mean())  # Filling null values with mean
print("\nDataFrame after filling null values with mean:")
display(df_filled_mean)

# 4. Forward fill or backward fill null values
df_ffill = df.ffill()  # Forward fill
df_bfill = df.bfill()  # Backward fill

# Displaying filled DataFrames
print("\nDataFrame after forward fill:")
display(df_ffill)
print("\nDataFrame after backward fill:")
display(df_bfill)

# EDA: Visualizing data distributions and relationships
plt.figure(figsize=(14, 10))

# Histograms of numerical columns
plt.subplot(3, 2, 1)
plt.hist(df['A'].dropna(), bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Column A')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(3, 2, 2)
plt.hist(df['C'].dropna(), bins=10, edgecolor='black', alpha=0.7, color='orange')
plt.title('Histogram of Column C')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Boxplots of numerical columns
plt.subplot(3, 2, 3)
sns.boxplot(y='D', data=df.dropna(), color='green')
plt.title('Boxplot of Column D')

plt.subplot(3, 2, 4)
sns.boxplot(y='E', data=df.dropna(), color='purple')
plt.title('Boxplot of Column E')

# Bar plot of categorical column
plt.subplot(3, 2, 5)
sns.countplot(x='B', data=df.dropna(), palette='pastel')
plt.title('Countplot of Column B')
plt.xticks(rotation=45)

# Scatter plot with numerical columns
plt.subplot(3, 2, 6)
sns.scatterplot(x='A', y='C', data=df.dropna(), hue='B', palette='Set1')
plt.title('Scatter Plot of Columns A vs C')
plt.xlabel('A')
plt.ylabel('C')

plt.tight_layout()
plt.show()

# Data manipulation example: Adding a new calculated column
df['F'] = df['A'] + df['C']
print("\nDataFrame after adding a new calculated column F:")
display(df)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a sample DataFrame with null values
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': ['a', 'b', 'c', np.nan, 'e'],
    'C': [np.nan, 'x', 'y', 'z', np.nan],
    'D': [10, 20, 30, 40, 50],
    'E': [np.nan, np.nan, np.nan, np.nan, np.nan]
}
df = pd.DataFrame(data)

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Method 1: Filling null values with a specific value
df_filled = df.fillna('Unknown')

# Print the DataFrame after filling null values
print("\nDataFrame after filling null values with 'Unknown':")
print(df_filled)

# Method 2: Filling null values with mean or median
mean_value = df['A'].mean()
df_mean_filled = df.fillna({'A': mean_value, 'C': df['C'].mode()[0]})

# Print the DataFrame after filling null values with mean and mode
print("\nDataFrame after filling null values with mean and mode:")
print(df_mean_filled)

# Method 3: Forward fill or backward fill
df_ffill = df.fillna(method='ffill')
df_bfill = df.fillna(method='bfill')

# Print the DataFrames after forward fill and backward fill
print("\nDataFrame after forward fill:")
print(df_ffill)
print("\nDataFrame after backward fill:")
print(df_bfill)

# Method 4: Interpolation
df_interpolated = df.interpolate(method='linear', limit_direction='both')

# Print the DataFrame after interpolation
print("\nDataFrame after interpolation:")
print(df_interpolated)

# Method 5: Drop rows or columns with null values
df_drop_rows = df.dropna()
df_drop_columns = df.dropna(axis=1)

# Print the DataFrames after dropping rows and columns with null values
print("\nDataFrame after dropping rows with null values:")
print(df_drop_rows)
print("\nDataFrame after dropping columns with null values:")
print(df_drop_columns)

# Method 6: Replace null values based on conditions
df_replace = df.copy()
df_replace['E'] = df_replace['E'].replace(np.nan, 'No Data')

# Print the DataFrame after replacing null values based on conditions
print("\nDataFrame after replacing null values based on conditions:")
print(df_replace)

# Method 7: Using isnull() and notnull() functions
null_mask = df['E'].isnull()
not_null_mask = df['E'].notnull()

# Print the masks and filtered DataFrame
print("\nNull mask for column 'E':")
print(null_mask)
print("\nNot null mask for column 'E':")
print(not_null_mask)
print("\nFiltered DataFrame where column 'E' is not null:")
print(df[not_null_mask])

# Method 8: Using apply() function to handle null values
def handle_nulls(column):
    if pd.isnull(column):
        return 'Unknown'
    else:
        return column

df_apply = df.applymap(handle_nulls)

# Print the DataFrame after applying handle_nulls function
print("\nDataFrame after applying function to handle null values:")
print(df_apply)

# Method 9: Using assert statement to check for null values
assert df['D'].notnull().all(), "There are null values in column 'D'"

# Method 10: Using fillna() with method='pad' to forward fill
df_pad = df.fillna(method='pad')

# Print the DataFrame after forward filling with pad method
print("\nDataFrame after forward filling with pad method:")
print(df_pad)

# Generate additional lines for demonstration
for i in range(10):
    print(f"\nAdditional line {i+1} for demonstration purposes.")

# End of code
print("\nEnd of code.")
```
