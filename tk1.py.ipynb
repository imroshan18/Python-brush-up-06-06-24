```python
import tkinter as tk
from tkinter import messagebox

def on_button_click():
    messagebox.showinfo("Button", "Button was clicked!")

def on_checkbutton_toggle():
    print("Checkbutton state:", check_var.get())

def on_radiobutton_select():
    print("Selected radiobutton:", radio_var.get())

def on_listbox_select(event):
    selected_items = [listbox.get(i) for i in listbox.curselection()]
    print("Selected items:", selected_items)

# Create the main application window
root = tk.Tk()
root.title("Tkinter Widgets Example")

# Create a Frame to hold all widgets
main_frame = tk.Frame(root, padx=10, pady=10)
main_frame.pack(fill="both", expand=True)

# Label
label = tk.Label(main_frame, text="This is a Label")
label.pack(pady=5)

# Entry
entry = tk.Entry(main_frame, width=30)
entry.pack(pady=5)

# Button
button = tk.Button(main_frame, text="Click Me", command=on_button_click)
button.pack(pady=5)

# Checkbutton
check_var = tk.BooleanVar()
checkbutton = tk.Checkbutton(main_frame, text="Check Me", variable=check_var, command=on_checkbutton_toggle)
checkbutton.pack(pady=5)

# Radiobuttons
radio_var = tk.StringVar(value="Option 1")
radiobutton_frame = tk.Frame(main_frame)
radiobutton_frame.pack(pady=5)

radiobutton1 = tk.Radiobutton(radiobutton_frame, text="Option 1", variable=radio_var, value="Option 1", command=on_radiobutton_select)
radiobutton2 = tk.Radiobutton(radiobutton_frame, text="Option 2", variable=radio_var, value="Option 2", command=on_radiobutton_select)

radiobutton1.pack(side="left")
radiobutton2.pack(side="left")

# Listbox with Scrollbar
listbox_frame = tk.Frame(main_frame)
listbox_frame.pack(pady=5)

scrollbar = tk.Scrollbar(listbox_frame, orient="vertical")
listbox = tk.Listbox(listbox_frame, selectmode="multiple", yscrollcommand=scrollbar.set, height=5)
scrollbar.config(command=listbox.yview)

for item in range(1, 21):
    listbox.insert(tk.END, f"Item {item}")

scrollbar.pack(side="right", fill="y")
listbox.pack(side="left", fill="both", expand=True)
listbox.bind("<<ListboxSelect>>", on_listbox_select)

# Canvas
canvas = tk.Canvas(main_frame, width=200, height=100, bg="white")
canvas.pack(pady=5)
canvas.create_line(0, 0, 200, 100, fill="blue")
canvas.create_rectangle(50, 25, 150, 75, fill="green")

# Menu
menubar = tk.Menu(root)
file_menu = tk.Menu(menubar, tearoff=0)
file_menu.add_command(label="Open")
file_menu.add_command(label="Save")
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)
menubar.add_cascade(label="File", menu=file_menu)

root.config(menu=menubar)

# Start the main event loop
root.mainloop()
```
