```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Generating some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a figure with multiple subplots
plt.figure(figsize=(12, 8))

# Line plot
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='green', linestyle='--', linewidth=2)
plt.title('Line Plot of sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Scatter plot
plt.subplot(2, 2, 2)
plt.scatter(x, y1, label='sin(x)', color='red', marker='o')
plt.scatter(x, y2, label='cos(x)', color='purple', marker='x')
plt.title('Scatter Plot of sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [25, 50, 75, 100]
plt.subplot(2, 2, 3)
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'])
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)

# Histogram
data = np.random.normal(0, 1, 1000)
plt.subplot(2, 2, 4)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Adjusting layout and displaying the plots
plt.tight_layout()
plt.show()
```
